```python
input_file = '../../data/maps/LittlerootTown/scripts.inc'
input_text = None
with open(input_file, 'r', encoding = 'utf-8') as f:
    input_text = f.read()

input_text
```

```python
all_lines = [line.strip() for line in input_text.split('\n') if line != '']
all_lines
```

```python
scripts = []
current_script = []
for line in all_lines:
    if line[0] == '@':
        # Full-line comments mess with the script identification process,
        # which uses the last line in the script.
        continue
    if ':' in line and '.string' not in line:
        # New function.
        if len(current_script) > 0:
            scripts.append(current_script)
        current_script = [line]
    else:
        current_script.append(line)
if len(current_script) > 0:
    scripts.append(current_script)

scripts
```

```python
tableScripts = []
eventScripts = []
textScripts = []
movementScripts = []

for script in scripts:
    first_tokens = script[0].split(':')
    name = first_tokens[0].strip()
    address = first_tokens[-1].replace('@', '').strip()
    
    destination = None
    if script[1].startswith('.string'):
        destination = textScripts
    elif script[-1] == 'step_end':
        destination = movementScripts
    elif script[-1].startswith('.byte') or script[-1].startswith('.2byte'):
        destination = tableScripts
    else:
        destination = eventScripts
    
    destination.append({
        'name': name,
        'address': address,
        'body': script[1:] 
    })
print("tables:")
display(tableScripts)
print("movement:")
display(movementScripts)
print("text:")
display(textScripts)
print("event:")
display(eventScripts)
```

```python
import copy

strippedScripts = copy.deepcopy(eventScripts)

def strip_comments(s):
    start = s.find('@')
    if start == -1:
        return s
    else:
        return s[:start].strip()

for script in strippedScripts:
    script['body'] = [strip_comments(line) for line in script['body'] if not line[0] == '@']

strippedScripts
```

```python
tokenizedScripts = copy.deepcopy(strippedScripts)

def make_tokens(s):
    tokens = []
    index = 0
    while index < len(s):
        if s[index].isalpha() or s[index] is '_':
            start_index = index
            index += 1
            while index < len(s) and (s[index].isalpha() or s[index].isdigit() or s[index] is '_'):
                index += 1
            tokens.append(s[start_index:index])
        elif s[index].isdigit():
            start_index = index
            index += 1
            if index < len(s) and s[index] == 'x':
                while index < len(s) and (s[index].isdigit() or s[index] in ['abcdefABCDEF']):
                    index += 1
            else:
                while index < len(s) and s[index].isdigit():
                    index += 1
            tokens.append(s[start_index:index])
        elif s[index].isspace():
            index += 1
        else:
            tokens.append(s[index])
            index += 1
    return tokens

for script in tokenizedScripts:
    script['body'] = [make_tokens(line) for line in script['body']]
    
tokenizedScripts
```

```python
parsedScripts = copy.deepcopy(tokenizedScripts)

LINKER_OPEN = '{'
LINKER_CLOSE = '}'

simple_mapping = {
    'end': 'return',
    'nop': '--# nop',
    'nop2': '--# nop2',
    'return': 'return',
    'killscript': 'Kill()'
}

compare_map = {
    'lt': '<',
    'eq': '==',
    'gt': '>',
    'le': '<=',
    'ge': '>=',
    'ne': '!='
}


def parse_script(script):
    index = 0
    named_scripts = []
    full_lua_text = ""
    while index < len(script['body']):
        lua_output = None
        
        line = script['body'][index]
        if line[0] == 'goto':
            assert(len(line) == 2)
            dest = line[1]
            named_scripts.append(dest)
            lua_output = LINKER_OPEN + 'goto ' + dest + LINKER_CLOSE
        elif line[0] == 'call':
            assert(len(line) == 2)
            dest = line[1]
            named_scripts.append(dest)
            lua_output = LINKER_OPEN + 'call ' + dest + LINKER_CLOSE
        elif line[0] in simple_mapping:
            assert(len(line) == 1)
            lua_output = simple_mapping[line[0]]
        elif line[0] == 'special':
            assert(len(line) == 2)
            func = line[1]
            lua_output = 'special.' + func + '()'
        elif line[0] == 'gotonative':
            assert(len(line) == 2)
            func = line[1]
            lua_output = 'native.' + func + '()\nreturn'
        elif line[0] == 'specialvar':
            assert(len(line) == 4)
            dest = line[1]
            assert(line[2] == ',')
            source = line[3]
            lua_output = 'special_vars.' + dest + ' = ' + source
        elif line[0] == 'compare':
            assert(len(line) == 4)
            first = line[1]
            assert(line[2] == ',')
            second = line[3]
            index += 1
            
            line = script['body'][index]
            assert(len(line) == 2)
            dest = line[1]
            named_scripts.append(dest)
            subtokens = line[0].split('_')
            assert(len(subtokens) == 3)
            assert(subtokens[1] == 'if')
            assert(subtokens[2] in compare_map)
            assert(subtokens[0] == 'call' or subtokens[0] == 'goto')
            op = compare_map[subtokens[2]]
            
            lua_output = 'if ' + first + ' ' + op + ' ' + second + '\nthen\n\t'
            lua_output += LINKER_OPEN + subtokens[0] + ' ' + dest + LINKER_CLOSE
            lua_output += '\nend'
        elif line[0] == 'setflag':
            assert(len(line) == 2)
            lua_output = 'flags.' + line[1] + ' = true'
        elif len(line[0]) > 7 and (line[0][0:4] == 'call' or line[0][0:4] == 'goto') and line[0][-3:] == 'set':
            assert(len(line) == 4)
            dest = line[3]
            named_scripts.append(dest)
            flag = line[1]
            
            subtokens = line[0].split('_')
            assert(len(subtokens) == 3)
            assert(subtokens[1] == 'if')
            assert(subtokens[0] == 'call' or subtokens[0] == 'goto')
            maybe_not = None
            if subtokens[2] == 'set':
                maybe_not = ''
            elif subtokens[2] == 'unset':
                maybe_not = 'not ' # Note: space required
            else:
                raise Exception('goto_if/call_if variant "' + line[0] + '" in ' + script['name'])
                
            lua_output = 'if ' + maybe_not + 'flags.' + flag + '\nthen\n\t'
            lua_output += LINKER_OPEN + subtokens[0] + ' ' + dest + LINKER_CLOSE
            lua_output += '\nend'
        elif line[0] == 'setvar':
            assert(len(line) == 4)
            dest = line[1]
            assert(line[2] == ',')
            value = line[3]
            lua_output = 'vars.' + dest + ' = ' + value
        elif line[0] == 'setflag':
            assert(len(line) == 2)
            lua_output = 'flags.' + line[1] + ' = true'
        elif line[0] == 'clearflag':
            assert(len(line) == 2)
            lua_output = 'flags.' + line[1] + ' = false'
        elif line[0] in encode_as_func:
            lua_output = line[0] + '('
            for token in line[1:]:
                lua_output += token
                if token == ',':
                    lua_output += ' '
            lua_output += ')'
        else:
            raise Exception('unknown script command "' + line[0] + '" in ' + script['name'])
        
        full_lua_text += lua_output + '\n'
        index += 1
    script['lua'] = full_lua_text
    script['references'] = named_scripts


encode_as_func = [
    'setobjectxyperm',
    'setobjectmovementtype',
    'checkplayergender',
    'lockall',
    'warpsilent',
    'waitstate',
    'releaseall',
    'delay',
    'playse',
    'applymovement',
    'waitmovement',
    'opendoor',
    'waitdooranim',
    'addobject',
    'closedoor',
    'msgbox',
    'closemessage',
    'hideobjectat',
    'warp',
    'setobjectxy',
    'lock',
    'faceplayer',
    'release',
    'removeobject',
    'playfanfare',
    'message',
    'waitfanfare'
]
    
for script in parsedScripts:
    parse_script(script)

parsedScripts
```
